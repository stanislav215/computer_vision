```
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, exposure, transform,io
from mpl_toolkits.mplot3d import axes3d
from os import path
```

```
coins = data.coins()
chelsea = data.chelsea()
coffee = data.coffee()
width = 210
images = [coins,chelsea,coffee]

for i in range(0,len(images)):
    dim = images[i].shape[0:2]
    images[i] = transform.resize(images[i],np.array((int(width*dim[0]/dim[1]),width)),order=4)

coins = images[0]
chelsea = images[1]
coffee = images[2]
```

```
original = coins
img_eq = exposure.equalize_hist(coins)
img_adapt_eq = exposure.equalize_adapthist(coins)
img_contrast_enhanced = exposure.adjust_gamma(coins, 2)
```

```
plt.figure(figsize=(10, 8))
ax1=plt.subplot(221)
ax1.set_title("Original image")
ax1.imshow(original,cmap='gray')

ax2 = plt.subplot(2,2,2)
ax2.imshow(img_contrast_enhanced,cmap='gray')
ax2.set_title("Imadjust")


ax3 = plt.subplot(2,2,3)
ax3.imshow(img_eq,cmap='gray')
ax3.set_title("Histeq")


ax4 = plt.subplot(2,2,4)
ax4.imshow(img_adapt_eq,cmap='gray')
ax4.set_title("Adapthisteq")
plt.show()
```

```
original = chelsea[:,:,0]
img_eq = exposure.equalize_hist(original)
img_adapt_eq = exposure.equalize_adapthist(original)
img_contrast_enhanced = exposure.adjust_gamma(original, 2)
```
